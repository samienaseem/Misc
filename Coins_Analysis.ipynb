```python
import ccxt
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Initialize the exchange (e.g., Binance)
exchange = ccxt.kraken()

# Fetch all tickers
tickers = exchange.fetch_tickers()
```

```python

# Helper function to get the historical data of a ticker
def get_historical_data(symbol):
    # Get the historical data for 1 year
    since = exchange.parse8601('2023-06-23T00:00:00Z')  # adjust the date to get the past year data
    ohlcv = exchange.fetch_ohlcv(symbol, '1d', since)
    df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
    return df

# List to store coins that meet the criteria
eligible_coins = []

# Loop through the tickers
for symbol in tickers:
    try:
        # Fetch ticker details
        ticker = tickers[symbol]
        info = ticker['info']
        # Check if the coin is at least 5 years old
        listed_date = datetime.strptime(info.get('listedDate', '1970-01-01'), '%Y-%m-%d')
        if datetime.now() - listed_date >= timedelta(days=5*365):
            # Get historical data
            df = get_historical_data(symbol)
            # Calculate the 52-week high
            high_52_week = df['high'].max()
            # Check if the current price is at 52-week high
            if ticker['last'] >= high_52_week:
                eligible_coins.append(symbol)
    except Exception as e:
        # Handle exceptions (e.g., missing data)
        print(f"Error processing {symbol}: {e}")

# Print the eligible coins
print("Coins that broke 52-week high and are at least 5 years old:")
for coin in eligible_coins:
    print(coin)
```

```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Fetch the ticker data from WazirX
def fetch_ticker_data():
    response = requests.get('https://api.wazirx.com/api/v2/tickers')
    return response.json()

# Fetch historical data for a specific symbol
def fetch_historical_data(symbol):
    end_time = int(datetime.now().timestamp())
    start_time = int((datetime.now() - timedelta(days=365)).timestamp())
    url = f'https://api.wazirx.com/api/v2/ohlc?symbol={symbol}&interval=1d&start={start_time}&end={end_time}'
    response = requests.get(url)
    data = response.json().get('data', {}).get(symbol, [])
    df = pd.DataFrame(data, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
    return df

# Check if the coin is at least 3 years old
def is_old_enough(listing_date):
    listing_date = datetime.strptime(listing_date, '%Y-%m-%d')
    return datetime.now() - listing_date >= timedelta(days=3*365)

# Main function to find coins that broke their 52-week high
def find_coins_breaking_52_week_high():
    tickers = fetch_ticker_data()
    eligible_coins = []

    for symbol, data in tickers.items():
        try:
            if 'baseAsset' in data and 'quoteAsset' in data:
                # Check if the coin is at least 3 years old
                base_asset = data['baseAsset']
                quote_asset = data['quoteAsset']
                symbol_info = f"{base_asset}{quote_asset}".lower()
                listing_date = data.get('at', '1970-01-01')
                if is_old_enough(listing_date):
                    # Fetch historical data
                    df = fetch_historical_data(symbol_info)
                    if df.empty:
                        continue
                    # Calculate the 52-week high
                    high_52_week = df['high'].max()
                    # Check if the current price is at the 52-week high
                    current_price = float(data['last'])
                    if current_price >= high_52_week:
                        eligible_coins.append(symbol)
        except Exception as e:
            print(f"Error processing {symbol}: {e}")

    return eligible_coins

# Find and print eligible coins
eligible_coins = find_coins_breaking_52_week_high()
print("Coins that broke 52-week high and are at least 3 years old:")
for coin in eligible_coins:
    print(coin)
```
